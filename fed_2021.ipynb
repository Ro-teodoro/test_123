```python
#agrupamos los campos por el nuevo campo calculado usando la suma paara los votos de cada aprtido
grouped = pd.DataFrame(data.groupby('ident').agg({
'CLAVE_CASILLA': 'first',
'CLAVE_ACTA': 'first',
'ID_ESTADO': 'first',
'NOMBRE_ESTADO': 'first',
'ID_DISTRITO': 'first',
'NOMBRE_DISTRITO': 'first',
'SECCION': 'first',
'ID_CASILLA': 'first',
'TIPO_CASILLA': 'first',
'EXT_CONTIGUA': 'first',
'CASILLA': 'first',
'NUM_ACTA_IMPRESO': 'first',
'PAN': 'sum',
'PRI': 'sum',
'PRD': 'sum',
'PVEM': 'sum',
'PT': 'sum',
'MC': 'sum',
'MORENA': 'sum',
'PES': 'sum',
'RSP': 'sum',
'FXM': 'sum',
'CI': 'sum',
'PAN-PRI-PRD': 'sum',
'PAN-PRI': 'sum',
'PAN-PRD': 'sum',
'PRI-PRD': 'sum',
'PVEM-PT-MORENA': 'sum',
'PVEM-PT': 'sum',
'PVEM-MORENA': 'sum',
'PT-MORENA': 'sum',
'CANDIDATO/A NO REGISTRADO/A': 'sum',
'VOTOS NULOS': 'sum',
'TOTAL_VOTOS_CALCULADOS': 'sum',
'LISTA_NOMINAL_CASILLA': 'sum',
'OBSERVACIONES': 'first',
'MECANISMOS_TRASLADO': 'first',
'FECHA_HORA': 'first'
}))

to_drop =['CLAVE_CASILLA', #columnas a eliminar
'CLAVE_ACTA',
'ID_ESTADO',
'NOMBRE_ESTADO',
'ID_DISTRITO',
'NOMBRE_DISTRITO',
'SECCION',
'ID_CASILLA',
'TIPO_CASILLA',
'EXT_CONTIGUA',
'CASILLA',
'NUM_ACTA_IMPRESO',
'OBSERVACIONES',
'MECANISMOS_TRASLADO',
'FECHA_HORA']

#eliminamos campos que ya no se usaran el la agrupacion
grouped = grouped.drop(to_drop, axis=1)
grouped.sample(5)
```

```python
col_to_sum = ['PAN', 'PRI', 'PRD', 'PVEM', 'PT', 'MC', 'MORENA', 'PES', 'RSP', 'FXM',
       'CI', 'PAN-PRI-PRD', 'PVEM-PT-MORENA', 'CANDIDATO/A NO REGISTRADO/A']

for i in col_to_sum:
    grouped[i+'_sum'] = 0

grouped.head(5)
```

```python
#agrupamos la tabla por candidato.
grouped_coal = dip.groupby('CANDIDATURA_PROPIETARIA',as_index=False).agg({
'PARTIDO_CI': ' | '.join, #se concatena
'ident': 'first'
}).reset_index()

#agrupamos la tabla por distrito.
grouped_coal = dip.groupby('ident',as_index=False).agg({
'PARTIDO_CI': ' | '.join, #se concatena
'ident': 'first'
}).reset_index()
```

```python
grouped_coal.head(8)
```

```python
grouped_coal.head(8)['PARTIDO_CI'][0]
```

```python

grouped_coal['coal'] = grouped_coal['PARTIDO_CI'].str.split("|").apply(lambda x: [i.strip() for i in x]) #agregamos una columna con la lista de partidos
grouped_coal = grouped_coal.drop(['PARTIDO_CI'], axis=1) #aliminamos  campos  que no nos sirven
grouped_coal['C-PAN-PRI-PRD'] = ''
grouped_coal['C-PVEM-PT-MORENA'] = ''
grouped_coal['C-indep'] = ''

print(grouped_coal['ident'].nunique())
print(grouped_coal.shape)
grouped_coal.sample(5)
```

```python
grouped_coal.head(8)['coal'][0][10]
```

```python
#aqui revisamos cuales distritos tienen coalicion y cuales no.
for i in grouped_coal.index:
    
    if 'PAN-PRI-PRD' in grouped_coal.loc[i,'coal']: 
        grouped_coal.loc[i,'C-PAN-PRI-PRD'] = 'PAN-PRI-PRD'
    else:
        grouped_coal.loc[i,'C-PAN-PRI-PRD'] = 'no'

    if 'PVEM-PT-MORENA' in grouped_coal.loc[i,'coal']: 
        grouped_coal.loc[i,'C-PVEM-PT-MORENA'] = 'PVEM-PT-MORENA'
    else: 
        grouped_coal.loc[i,'C-PVEM-PT-MORENA'] = 'no'

    if (('CI-1' in grouped_coal.loc[i,'coal']) or ('CI-2' in grouped_coal.loc[i,'coal']) or ('CI-3' in grouped_coal.loc[i,'coal'])): 
        grouped_coal.loc[i,'C-indep'] = 'si'
    else: 
        grouped_coal.loc[i,'C-indep'] = 'no'


grouped_coal.sample(5)
```

```python
result = pd.merge(grouped, grouped_coal[['ident','C-PAN-PRI-PRD','C-PVEM-PT-MORENA','C-indep' ]], on='ident', how='left')
print('tabla de resultados',grouped.shape)

print('tabla de coaliciones',grouped_coal.shape)

print('tabla resultado',result.shape)
result.sample(5)
```

```python
result2 = pd.merge(data, grouped_coal[['ident','C-PAN-PRI-PRD','C-PVEM-PT-MORENA','C-indep' ]], on='ident', how='left')
print('tabla de datos',data.shape)

print('tabla de coaliciones',grouped_coal.shape)

print('tabla resultado',result2.shape)
result2.sample(5)
```

```python
col_to_sum = ['PAN', 'PRI', 'PRD', 'PVEM', 'PT', 'MC', 'MORENA', 'PES', 'RSP', 'FXM',
       'CI', 'PAN-PRI-PRD', 'PVEM-PT-MORENA', 'CANDIDATO/A NO REGISTRADO/A']

for i in col_to_sum:
    result2[i+'_sum'] = 0

result2.head(5)
```

```python
result2['PAN_sum'] = result2['PAN']
result2['PRI_sum'] = result2['PRI']
result2['PRD_sum'] = result2['PRD']
result2['PVEM_sum'] = result2['PVEM']
result2['PT_sum'] = result2['PT'] 
result2['MORENA_sum'] = result2['MORENA']
result2['MC_sum'] = result2['MC']
result2['PES_sum'] = result2['PES']
result2['RSP_sum'] = result2['RSP']
result2['FXM_sum'] = result2['FXM']
result2['CI_sum'] = result2['CI']
result2['CANDIDATO/A NO REGISTRADO/A_sum'] = result2['CANDIDATO/A NO REGISTRADO/A']
result2['PAN-PRI-PRD_sum'] =  result2['PAN']+ result2['PRI'] + result2['PRD'] + result2['PAN-PRI']+ result2['PAN-PRD']+ result2['PRI-PRD']
result2['PVEM-PT-MORENA_sum'] = result2['PVEM']+ result2['PT'] + result2['MORENA'] + result2['PVEM-PT']+ result2['PVEM-MORENA']+ result2['PT-MORENA']
result2.sample(5)
```

```python
result2.shape
```

```python
for i in result2.index:
    if result2.loc[i,'C-PAN-PRI-PRD']=='PAN-PRI-PRD':
        result2.loc[i, ['PAN_sum', 'PRI_sum', 'PRD_sum']] = 0
    else:
        result2.loc[i, ['PAN_sum', 'PRI_sum', 'PRD_sum']] = 10
```

```python
mask1 = result2['C-PAN-PRI-PRD'] == 'PAN-PRI-PRD'
result2.loc[mask1, ['PAN_sum', 'PRI_sum', 'PRD_sum']] = 0
result2.loc[~mask1, 'PAN-PRI-PRD_sum'] = 0

mask2 = result2['C-PVEM-PT-MORENA'] == 'PVEM-PT-MORENA'
result2.loc[mask2, ['PVEM_sum', 'PT_sum', 'MORENA_sum']] = 0
result2.loc[~mask2, 'PVEM-PT-MORENA_sum'] = 0
result2.sample(5)
```

```python

#agrupamos la tabla por seccion.
by_seccion = result2.groupby(['ID_ESTADO', 'ID_DISTRITO' , 'SECCION'],as_index=False).agg({'ID_ESTADO' : 'first', 
'NOMBRE_ESTADO': 'first', 
'ID_DISTRITO': 'first', 
'NOMBRE_DISTRITO': 'first',
'SECCION': 'first', 
'VOTOS NULOS' : 'sum',
'TOTAL_VOTOS_CALCULADOS': 'sum', 
'LISTA_NOMINAL_CASILLA': 'sum', 
'ident': 'first', 
'PAN_sum': 'sum', 
'PRI_sum': 'sum', 
'PRD_sum': 'sum', 
'PVEM_sum': 'sum', 
'PT_sum': 'sum', 
'MC_sum': 'sum', 
'MORENA_sum': 'sum', 
'PES_sum': 'sum', 
'RSP_sum': 'sum', 
'FXM_sum': 'sum', 
'CI_sum': 'sum', 
'PAN-PRI-PRD_sum': 'sum', 
'PVEM-PT-MORENA_sum': 'sum', 
'CANDIDATO/A NO REGISTRADO/A_sum': 'sum' }).reset_index()

by_seccion.sample(6)
```

```python
cols_w_votes = [ 'PAN_sum',
       'PRI_sum', 'PRD_sum', 'PVEM_sum', 'PT_sum', 'MC_sum', 'MORENA_sum', 'PES_sum', 'RSP_sum', 'FXM_sum', 'CI_sum',
       'PAN-PRI-PRD_sum', 'PVEM-PT-MORENA_sum', 'CANDIDATO/A NO REGISTRADO/A_sum']
       
by_seccion['max_col'] = by_seccion[cols_w_votes].idxmax(axis=1) #esta funcion agrega una columna anunciando cual tiene la mayor cantidad de votos
by_seccion.sample(10)
```

```python
by_seccion['max_col'] = by_seccion['max_col'].str.replace('_sum', '')
by_seccion['nident'] = by_seccion['ID_ESTADO'] + by_seccion['SECCION']
by_seccion.sample(5)
```
